```python
from datasets import load_dataset
```

```python
ds = load_dataset("legacy-datasets/banking77")
```

```python
label_names = ds["train"].features["label"].names

print(sorted(label_names))
```

```python
import re

def clean_intent(name):
    name = re.sub(r'[?]', '', name)
    name = name.lower()
    return name

cleaned = sorted([clean_intent(n) for n in label_names])
print(cleaned)
```
